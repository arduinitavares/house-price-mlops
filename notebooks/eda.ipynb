```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de display
pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")
```

```python
file_path = "data/raw/AmesHousing.csv"
df = pd.read_csv(file_path)
df.shape
```

```python
df.head()
```

```python
df.info()
```

## Verificar valores ausentes

```python
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

## Estatísticas descritivas numéricas

```python
df.describe()
```

## Correlação com variável target

```python
target = "SalePrice"
corr = df.corr(numeric_only=True)
top_corr = corr[target].sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_corr.values, y=top_corr.index)
plt.title("Top 10 correlações com SalePrice")
plt.tight_layout()
plt.show()
```

## Heatmap de correlação

```python
# Cell 8: Heatmap de correlação
plt.figure(figsize=(12, 10))
sns.heatmap(corr[top_corr.index].loc[top_corr.index], annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlação entre variáveis mais correlacionadas com SalePrice")
plt.tight_layout()
plt.show()
```
